```python
df.isnull().sum()
print(df.groupby("sex")["body_mass_g"].mean())
print(df.groupby("sex")["flipper_length_mm"].mean())
```

```python
import numpy as np
gender_selection= np.where(df["body_mass_g"]>4700,"Male","Female")
df["sex"].fillna(pd.Series(gender_selection),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["bill_length_mm"].fillna(df.groupby("sex")["bill_length_mm"].transform("mean"),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["bill_depth_mm"].fillna(df.groupby("sex")["bill_depth_mm"].transform("mean"),inplace=True)
```

```python
df.isnull().sum()
```

```python
df["flipper_length_mm"].fillna(df.groupby("sex")["flipper_length_mm"].transform("mean"),inplace=True)
```

```python
df["body_mass_g"].fillna(df.groupby("sex")["body_mass_g"].transform("mean"),inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
#relation between bodymass and flip length
mass_bill_length =df.groupby("bill_length_mm")["body_mass_g"].mean()
plt.plot(mass_bill_length.index,mass_bill_length.values,marker="X")
plt.ylabel("Bill length")
plt.xlabel("Body mass")
plt.title("Body mass vs flip length")
```

```python
mass_flip_depth =df.groupby("flipper_length_mm")["body_mass_g"].mean()
plt.plot(mass_flip_depth.index,mass_flip_depth.values,marker="X")
plt.ylabel("flip depth")
plt.xlabel("Body mass")
plt.title("Body mass vs flip depth")
```

```python
#island vs body mass
island_wise_body_mass=df.groupby("island")["body_mass_g"].mean()
plt.bar(island_wise_body_mass.index,island_wise_body_mass.values)
```

```python
#species wise body mass
species_wise_body_mass= df.groupby("species")["body_mass_g"].mean()
species_wise_body_mass
plt.bar(species_wise_body_mass.index,species_wise_body_mass.values,color="green")
xlabel="species"
ylabel="weight"
```
