```python
df.rename(columns={"Customer Id":"customer_id","Gender":"gender","Age":"age","City":"city","Membeship Type":"membership_type","Total Spend":"total_sped","Items Purchased":"items_purchased","Average Rating":"average_rating","Discount Applied":"discount_applied","Days Since Last Purchase":"days_since_last_purchase","Satisfaction Level":"satisfaction_level"},inplace=True)
```

```python
condition= df["average_rating"].apply(lambda x: "Unsatisfied" if x<3.5 else "Neutral" if x>3.5 and x<4.0 else "Satisfied"  )
df["satisfaction_level"].fillna(condition ,inplace=True)
```

```python
#membership wise count
membership_count= df["Membership Type"].value_counts()
plt.plot(membership_count.index,membership_count.values,color="green",marker="X",linestyle="none")
plt.xlabel("Memebership_type",size=15)
plt.ylabel("count",size=15)
plt.title("Membership Analysis",color= "RED",size="25")
```

```python
#membership wise total spend
membership_wise_spend =  df.groupby("Membership Type")["total_sped"].sum()
plt.bar(membership_wise_spend.index,membership_wise_spend.values)
plt.xlabel("Membershipo type")
plt.title("membership_wise_spend",size=20,color="green")
plt.ylabel("total spend")
```
