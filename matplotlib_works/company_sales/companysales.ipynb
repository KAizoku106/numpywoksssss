```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\LENOVO\\Desktop\\Development-jourey\\machinelearningworks\\matplotlib_works\\company_sales\\company_sales_data.csv")
df.head()
```

```python
#Read Total profit of all months and show it using a line plot
profit_of_months=df.groupby("month_number")["total_profit"].sum()
profit_of_months
plt.plot(profit_of_months.index,profit_of_months.values,color="red",marker=".",linestyle="dotted")
plt.xlabel("month number")
plt.ylabel("total_pofit")
plt.title("month vs total profit")
```

```python
#all_product_sales_data
all_product_sales_data=df.groupby("month_number")["total_units"].sum()
plt.plot(all_product_sales_data,color="red",linewidth=3,linestyle="dotted",marker="o")
plt.xlabel("month number")
plt.ylabel("total_unit_saled")
plt.title("month vs total unit saled")
```

## multiline plotting of each product sales


```python
plt.plot(df["month_number"],df["facecream"])
plt.plot(df["month_number"],df["toothpaste"])
plt.plot(df["month_number"],df["facewash"])
plt.plot(df["month_number"],df["shampoo"])
plt.plot(df["month_number"],df["moisturizer"])
```

```python
products=["facecream","toothpaste","facewash","shampoo","moisturizer"]
for p in products:
    plt.plot(df["month_number"],df[p],label=p)
plt.legend()    
```

## Read toothpaste sales data of each month and show it using a scatterplot

```python
toothpaste= df.groupby("month_number")["toothpaste"].sum()
toothpaste
```

```python
plt.scatter(df["month_number"],df["toothpaste"],label="unit_sold")
plt.legend()
plt.xlabel("month_number")
plt.ylabel("unit_solded")
plt.title("toothpaste sales in every month")
```

### Read face cream and facewash product sales data and show it using the bar chart with grid line

```python
bar_width=.35
plt.bar(df["month_number"]-bar_width/2,df["facecream"],width=bar_width,label="facecream")
plt.bar(df["month_number"]+bar_width/2,df["facewash"],width=bar_width,label="facewash")
plt.legend()
```

### Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.hist(df["total_profit"],bins=12,edgecolor="black")
```

### .Calculate total sale data for last year for each product and show it using a Pie chart



```python
products=["facecream","toothpaste","facewash","shampoo","moisturizer","bathingsoap"]

sales_summary ={p:df[p].sum() for p in products}
plt.pie(sales_summary.values(), labels=products,autopct="%1.1f%%")
plt.legend()
```

### Read Bathing soap facewash of all months and display it using the Subplot


```python
figure,axes= plt.subplots(1,2)
axes[0].plot(df["month_number"],df["bathingsoap"])
axes[0].set_title("bathingsoap")
axes[1].plot(df["month_number"],df["facewash"])
axes[1].set_title("facewash")
```
