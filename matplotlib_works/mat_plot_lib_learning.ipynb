```python
import  matplotlib.pyplot as plt
```

```python
food_logs  = [ "biriyani","tea","mandi","icecream"]
calories = [250,50,350,240]
calorieforchildrens= [200,30,200,200]
```

```python
plt .plot(food_logs,calories,marker="X",linestyle="dotted",color="black",linewidth=1)
plt .plot(food_logs,calorieforchildrens,marker="X",linestyle="dotted",color="red",linewidth=2)
plt.title("Foodlogs",color="Green",size="30")
plt.xlabel("foods",color="black")
plt.ylabel("calories",color="black")
plt.show()
```
